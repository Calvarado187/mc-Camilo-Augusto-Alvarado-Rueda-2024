```python
import numpy as np
import matplotlib.pyplot as plt

# Datos
x1 = np.array([1, 1, 2, 3, 1, 2, 3, 3])
x2 = np.array([0, 1, 1, 2, 2, 3, 3, 1])
y = np.array([0.6, 2, 0.1, 0.3, 2.2, 2.3, 0.8, -1])

# Ajuste a una función lineal: y = mx1 + nx2 + b
A = np.vstack([x1, x2, np.ones(len(x1))]).T
m, n, b = np.linalg.lstsq(A, y, rcond=None)[0]

# Coeficiente de correlación (r)
r = np.corrcoef([x1, x2, y])[0, 2]

# Graficar los datos y la línea de regresión
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x1, x2, y, c='r', marker='o')

# Calcular los valores de y para la línea de regresión
x1_range = np.linspace(min(x1), max(x1), 100)
x2_range = np.linspace(min(x2), max(x2), 100)
X1, X2 = np.meshgrid(x1_range, x2_range)
Y = m*X1 + n*X2 + b

# Graficar la línea de regresión
ax.plot_surface(X1, X2, Y, alpha=0.5)

ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('Y')

plt.title(f'Ajuste lineal con r = {r:.2f}')
plt.show()

print(f"Coeficiente de correlación (r): {r}")
print(f"Función lineal ajustada: y = {m:.2f}x1 + {n:.2f}x2 + {b:.2f}")
```
