```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange, CubicSpline

# Datos proporcionados
x = np.array([1, 2, 3, 4, 5, 6, 7])
y = np.array([1, 5, 4, 4, -2, 2, 9])

# InterpolaciÃ³n de Lagrange
polinomio_lagrange = lagrange(x, y)

# InterpolaciÃ³n con Trazadores CÃºbicos
cs = CubicSpline(x, y)

# Puntos para las grÃ¡ficas
x_vals = np.linspace(1, 7, 500)
y_lagrange = polinomio_lagrange(x_vals)
y_cubic_spline = cs(x_vals)

# EstimaciÃ³n del valor f(3.55)
valor_lagrange = polinomio_lagrange(3.55)
valor_cubic_spline = cs(3.55)

print(f"Valor estimado de f(3.55) usando Lagrange: {valor_lagrange}")
print(f"Valor estimado de f(3.55) usando Trazadores CÃºbicos: {valor_cubic_spline}")

# GrÃ¡ficas
plt.figure(figsize=(10, 6))

# GrÃ¡fica de Lagrange
plt.subplot(2, 1, 1)
plt.plot(x, y, 'o', label='Datos')
plt.plot(x_vals, y_lagrange, label='Lagrange')
plt.title('InterpolaciÃ³n de Lagrange')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()

# GrÃ¡fica de Trazadores CÃºbicos
plt.subplot(2, 1, 2)
plt.plot(x, y, 'o', label='Datos')
plt.plot(x_vals, y_cubic_spline, label='Trazadores CÃºbicos')
plt.title('InterpolaciÃ³n con Trazadores CÃºbicos')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()

plt.tight_layout()
plt.show()
```
