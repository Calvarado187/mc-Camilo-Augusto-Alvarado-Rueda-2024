```python
import numpy as np
import matplotlib.pyplot as plt

# Datos
x_values = np.array([0, 2, 4, 6, 8])
y_values = np.array([2, 0.1, 3, 4.5, 7])
target_x = 2.75

# Función para calcular los polinomios de interpolación de Lagrange
def lagrange_interpolation(x, y, target_x, degree):
    result = 0
    n = len(x)
    for i in range(n):
        term = y[i]
        for j in range(n):
            if i != j:
                term *= (target_x - x[j]) / (x[i] - x[j])
        result += term
    return result

# Grado 1
degree_1_result = lagrange_interpolation(x_values[:2], y_values[:2], target_x, 1)

# Grado 2
degree_2_result = lagrange_interpolation(x_values[:3], y_values[:3], target_x, 2)

# Grado 3
degree_3_result = lagrange_interpolation(x_values[:4], y_values[:4], target_x, 3)

print("Estimación de f(2.75) para grados 1, 2 y 3:")
print("Grado 1:", degree_1_result)
print("Grado 2:", degree_2_result)
print("Grado 3:", degree_3_result)

# Visualización de los resultados
plt.scatter(x_values, y_values, color='red', label='Datos originales')
plt.axvline(x=target_x, color='gray', linestyle='--', label='x = 2.75')
plt.plot(target_x, degree_1_result, 'bo', label='Grado 1')
plt.plot(target_x, degree_2_result, 'go', label='Grado 2')
plt.plot(target_x, degree_3_result, 'yo', label='Grado 3')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Interpolación de Lagrange')
plt.legend()
plt.grid(True)
plt.show()

```
